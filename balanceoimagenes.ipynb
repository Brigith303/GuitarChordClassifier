```python
import os
import random
import shutil
from PIL import Image, ImageEnhance

# ====== CONFIGURACI√ìN ======
DATASET_DIR = r"C:\Users\anton\Pictures\Tesis\GuitarSet"          # dataset original
OUTPUT_DIR  = r"C:\Users\anton\Pictures\Tesis\GuitarSet_Balanceado"  # nuevo dataset balanceado
TARGET_IMAGES = 100  # im√°genes deseadas por subcarpeta
EXTENSIONS = (".jpg", ".jpeg", ".png")

# ====== FUNCIONES DE AUMENTO ======
def augment_image(img):
    """Aplica una transformaci√≥n aleatoria suave a la imagen"""
    choice = random.choice(["rotate", "bright", "contrast", "flip"])
    
    if choice == "rotate":
        angle = random.uniform(-15, 15)
        img = img.rotate(angle)
    elif choice == "bright":
        enhancer = ImageEnhance.Brightness(img)
        img = enhancer.enhance(random.uniform(0.7, 1.3))
    elif choice == "contrast":
        enhancer = ImageEnhance.Contrast(img)
        img = enhancer.enhance(random.uniform(0.8, 1.2))
    elif choice == "flip":
        img = img.transpose(Image.FLIP_LEFT_RIGHT)
    
    return img

# ====== CREAR ESTRUCTURA DE SALIDA ======
if not os.path.exists(OUTPUT_DIR):
    os.makedirs(OUTPUT_DIR)

for chord in os.listdir(DATASET_DIR):
    chord_path = os.path.join(DATASET_DIR, chord)
    if not os.path.isdir(chord_path):
        continue
    
    for subfolder in os.listdir(chord_path):
        sub_path = os.path.join(chord_path, subfolder)
        if not os.path.isdir(sub_path):
            continue

        # Crear carpeta equivalente en el dataset balanceado
        new_sub_path = os.path.join(OUTPUT_DIR, chord, subfolder)
        os.makedirs(new_sub_path, exist_ok=True)

        # Obtener im√°genes
        images = [f for f in os.listdir(sub_path) if f.lower().endswith(EXTENSIONS)]
        count = len(images)

        print(f"\nüéµ {chord} - {subfolder}: {count} im√°genes encontradas")

        # Si hay m√°s de lo necesario ‚Üí muestreo aleatorio
        if count >= TARGET_IMAGES:
            print(f"   ‚öñÔ∏è Seleccionando 100 im√°genes aleatorias...")
            selected = random.sample(images, TARGET_IMAGES)
            for f in selected:
                src = os.path.join(sub_path, f)
                dst = os.path.join(new_sub_path, f)
                shutil.copy2(src, dst)

        # Si hay menos ‚Üí copiar todas y generar nuevas
        elif count > 0:
            print(f"   ‚ûï Generando {TARGET_IMAGES - count} im√°genes aumentadas...")
            # Copiar originales primero
            for f in images:
                src = os.path.join(sub_path, f)
                dst = os.path.join(new_sub_path, f)
                shutil.copy2(src, dst)
            
            # Crear im√°genes aumentadas
            for i in range(TARGET_IMAGES - count):
                img_name = random.choice(images)
                img_path = os.path.join(sub_path, img_name)
                img = Image.open(img_path).convert("RGB")
                aug = augment_image(img)
                new_name = f"aug_{i}_{img_name}"
                aug.save(os.path.join(new_sub_path, new_name))

        else:
            print(f"   ‚ö†Ô∏è Carpeta vac√≠a, se omitir√°.")

print("\n‚úÖ Dataset balanceado creado exitosamente en:")
print(f"üëâ {OUTPUT_DIR}")
```
